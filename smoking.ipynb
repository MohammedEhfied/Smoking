Let's see how many males and females in this dataframe

```python
female = []
for i in df.GENDER:
    if i == "F":
        female.append(i)        
print("There are",len(female),"women in this dataframe")
```

```python
male = []
for i in df.GENDER:
    if i == "M":
        male.append(i)
print("There are",len(male),"men in this dataframe")
```

How many smokers are in this dataframe?

```python
smokers = []
for i in df["SMOKING"]:
    if i == "YES":
        smokers.append(i)
print("There are",len(smokers),"smokers")   
```

```python
male_smokers = []
for i in df["ID"]:
    if df["GENDER"][i] == "M" and df["SMOKING"][i] == "YES":
        male_smokers.append(i)
female_smokers = []
for i in df["ID"]:
    if df["GENDER"][i] == "F" and df["SMOKING"][i] == "YES":
        female_smokers.append(i)
print(len(male_smokers),"of the smokers are men and",len(female_smokers),"are women")
```

```python
plt.pie([len(male_smokers),len(female_smokers)], labels=["male","female"],autopct='%.2f%%',explode =(0.05,0),
shadow=True,startangle=90)
plt.title('The ratio of male smokers to female smokers:')
plt.show()
```

How many smokers have a cough?

```python
cough = []
for i in df["ID"]:
    if df["COUGHING"][i] == "YES" and df["SMOKING"][i] == "YES":
      cough.append(i)
print("there are",len(cough),"smokers have a cough")
```

How many smokers have lung cancer?

```python
cancer = []
male_cancer = []
female_cancer = []
for i in df["ID"]:
    if df["LUNG_CANCER"][i] == "YES" and df["SMOKING"][i] == "YES":
        cancer.append(i)
for i in cancer:
    if df["GENDER"][i] == "M":
        male_cancer.append(i) 
for i in cancer:
    if df["GENDER"][i] == "F":
        female_cancer.append(i) 

print("there are",len(cancer),"smokers who have lung cancer","\n",len(male_cancer),"of them are men and",
len(female_cancer),"of them are women")
```
